### **Primo Grafico:** Incontro con Amici - Analisi Generale

La prima analisi del progetto riguarda il cambiamento nei rapporti sociali tra le diverse generazioni, con un focus specifico sulla frequenza degli incontri con amici. Questo studio evidenzia una tendenza alla diminuzione dei rapporti sociali, un fenomeno che è stato notevolmente accentuato dall'avvento della pandemia di COVID-19.

#### Aspetti Chiave dell'Analisi:
- **Diminuzione dei Rapporti Sociali:** Negli ultimi anni, si è osservata una riduzione costante nella frequenza degli incontri con amici tra le diverse fasce d'età. Questa tendenza ha subito un'accelerazione significativa durante e dopo la pandemia.
  
- **Impatto Negativo sulla Popolazione:** La riduzione dei rapporti sociali può avere numerosi effetti negativi sulla popolazione, influenzando vari aspetti della vita:
  - **Salute Mentale:** L'isolamento sociale può portare a sentimenti di solitudine, ansia e depressione.
  - **Qualità della Vita:** La mancanza di interazioni sociali può ridurre il senso di appartenenza e comunità, elementi fondamentali per il benessere individuale.
  - **Sviluppo Sociale dei Bambini:** Per i bambini e i giovani, le interazioni sociali sono cruciali per lo sviluppo delle competenze sociali e emotive.


```python
file_path = 'amici/incontroAmici3.csv'  

# Leggi il file CSV con il delimitatore corretto
df = pd.read_csv(file_path, delimiter=',', quotechar='"')

# Creiamo un istogramma
plt.figure(figsize=(9, 4))

# Usa Seaborn per creare un plot categorico
sns.barplot(
    data=df, 
    x='TIME', 
    y='Value', 
    hue='Tipo dato', 
    errorbar=None
)

# Aggiungi dettagli al grafico
plt.title("Percentuali per varie tipologie di incontri sociali per classi di età (2001, 2010, 2022)")
plt.xlabel("Anno")
plt.ylabel("Percentuale")
plt.legend(title="Tipo dato", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)

# Mostra il grafico
plt.show()
```

### **Secondo Grafico:** Incontro con Amici Suddiviso per Fasce d'Età

Il secondo grafico del progetto offre un'analisi dettagliata della frequenza degli incontri con amici, suddivisa per specifiche fasce d'età. Questo approfondimento è fondamentale per comprendere come i rapporti sociali variano non solo nel tempo, ma anche tra diverse generazioni.

#### Fasce d'Età Analizzate:
- ( **6-10 anni** ,  **11-14 anni** , **15-17 anni** , **18-19 anni**)
- ( **20-24 anni** , **25-34 anni**)
- ( **45-54 anni** , **55-59 anni** , **60-64 anni** , **65-74 anni** , **75 anni e più**)

#### Aspetti Chiave dell'Analisi:
- **Evoluzione dei Rapporti Sociali:** L'analisi per fasce d'età permette di osservare come le dinamiche sociali cambiano durante le diverse fasi della vita. Ad esempio, i bambini e gli adolescenti potrebbero avere frequenze di incontri con amici diverse rispetto agli adulti e agli anziani.
  
- **Impatto delle Età sulle Relazioni:** Con questa suddivisione, possiamo identificare le fasce d'età che hanno subito maggiormente la diminuzione dei rapporti sociali, specialmente durante periodi critici come la pandemia.
  
- **Trend Generazionali:** Esaminando le differenze tra le fasce d'età, possiamo comprendere meglio le tendenze generazionali e le possibili cause dei cambiamenti nei comportamenti sociali.

#### Obiettivi dell'Analisi:
- **Identificare le Fasce più Vulnerabili:** 


```python
# Definisce i gruppi di classi di età
gruppo1 = ['6-10 anni', '11-14 anni', '6 anni e più', '15-17 anni', '18-19 anni']
gruppo2 = ['20-24 anni', '25-34 anni']
gruppo3 = ['45-54 anni', '55-59 anni', '60-64 anni', '65-74 anni', '75 anni e più']

# Ottieni tutte le tipologie uniche di dati per garantire una palette coerente
tipi_dato_unici = df['Tipo dato'].unique()

# Crea una palette fissa con un colore per ogni tipologia unica di dati
palette = sns.color_palette("muted", n_colors=len(tipi_dato_unici))

# Crea un dizionario che associa i tipi di dato ai colori
palette_dict = dict(zip(tipi_dato_unici, palette))

# Funzione per creare gli istogrammi separati
def crea_istogramma(df, gruppi, titolo, palette_dict):
    plt.figure(figsize=(10, 2))
    df_group = df[df['Classe di età'].isin(gruppi)]
    sns.barplot(
        data=df_group, 
        x='TIME', 
        y='Value', 
        hue='Tipo dato', 
        errorbar=None,
        palette=palette_dict
    )
    plt.title(titolo)
    plt.xlabel("Anno")
    plt.ylabel("Percentuale")
    plt.legend(title="Tipo dato", bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(True)
    plt.show()

# Crea i tre istogrammi per i tre gruppi di classi di età
crea_istogramma(df, gruppo1, "Percentuali delle tipologie di incontri sociali (Gruppo 1: 6-10 anni, 11-14 anni, 6 anni e più, 15-17 anni, 18-19 anni)", palette_dict)
crea_istogramma(df, gruppo2, "Percentuali delle tipologie di incontri sociali (Gruppo 2: 20-24 anni, 25-34 anni)", palette_dict)
crea_istogramma(df, gruppo3, "Percentuali delle tipologie di incontri sociali (Gruppo 3: 45-54 anni, 55-59 anni, 60-64 anni, 65-74 anni, 75 anni e più)", palette_dict)
```

### **Terzo Grafico:** Partecipazione Sportiva

Il terzo grafico del progetto analizza la partecipazione sportiva in Italia, confrontando la distribuzione delle percentuali di partecipazione sportiva negli anni 2001, 2010 e 2022. Verranno creati tre grafici a torta, ciascuno rappresentante un anno specifico, per mostrare chiaramente le tendenze e i cambiamenti nella pratica sportiva nel corso del tempo.

#### Aspetti Chiave dell'Analisi:
- **Aumento della Partecipazione Sportiva Costante:** 
  - Si può osservare come la percentuale di persone che praticano sport con costanza sia in aumento nel tempo. Questo è un segnale positivo, dato che lo sport ha innumerevoli benefici per il corpo umano, tra cui miglioramento della salute fisica e mentale.
  
- **Benefici dello Sport:**
  - **Salute Fisica:** La pratica regolare dello sport contribuisce a mantenere il corpo in forma, prevenendo molte malattie croniche come l'obesità, il diabete e le malattie cardiache.
  - **Salute Mentale:** Lo sport è un ottimo modo per ridurre lo stress e migliorare il benessere emotivo, grazie alla produzione di endorfine durante l'attività fisica.
  - **Socializzazione:** Praticare sport spesso implica partecipare a gruppi o club, il che può migliorare le competenze sociali e creare nuove amicizie.

- **Stile di Vita Sedentario e Lavori Non Fisici:**
  - Con l'aumento dei lavori sedentari e non fisici, sempre più persone cercano lo sport come una via di fuga per scaricare lo stress accumulato durante la giornata lavorativa. Questo cambiamento culturale riflette una maggiore consapevolezza dell'importanza di mantenere uno stile di vita attivo per compensare gli effetti negativi di una vita sedentaria.

#### Obiettivi dell'Analisi:
- **Identificare i Cambiamenti nelle Abitudini Sportive:**
  - Analizzare i cambiamenti nella partecipazione sportiva nel corso del tempo permette di comprendere meglio le abitudini e le tendenze della popolazione italiana rispetto allo sport.

I grafici a torta forniranno una rappresentazione visiva immediata e chiara dei cambiamenti nella partecipazione sportiva, aiutando a mettere in evidenza l'evoluzione delle abitudini sportive in Italia.

```python
# Caricare i file CSV
sport2001 = pd.read_csv("sport/sport2001.csv")
sport2010 = pd.read_csv("sport/sport2010.csv")
sport2022 = pd.read_csv("sport/sport2022.csv")

# Filtrare i dati totali per anno
sport2001_totale = sport2001[sport2001['Sesso'] == 'totale']
sport2010_totale = sport2010[sport2010['Sesso'] == 'totale']
sport2022_totale = sport2022[sport2022['Sesso'] == 'totale']

# Creare un dizionario per i dati per ciascun anno
dati = {
    '2001': sport2001_totale.groupby('Tipo dato')['Value'].sum(),
    '2010': sport2010_totale.groupby('Tipo dato')['Value'].sum(),
    '2022': sport2022_totale.groupby('Tipo dato')['Value'].sum()
}

# Creare i grafici a torta
fig, axn = plt.subplots(1, 3, figsize=(18, 6))

for ax, (year, values) in zip(axn, dati.items()):
    labels = values.index
    ax.pie(values, labels=labels, autopct='%.1f%%', startangle=140)
    ax.set_title(f'Partecipazione Sportiva {year}')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Leggiamo i dati dal file CSV
df = pd.read_csv('vita/vita2022.csv')

# Convertiamo la colonna "Value" in numerico, in caso contenga dati non numerici
df['Value'] = pd.to_numeric(df['Value'], errors='coerce')

# Rimuoviamo eventuali righe con valori mancanti
df = df.dropna(subset=['Value'])

# Ordiniamo il DataFrame per la colonna 'Classe di età'
df = df.sort_values(by='Classe di età')

# Creiamo il grafico a linee
plt.figure(figsize=(14, 10))
line_plot = sns.lineplot(
    data=df,
    x='Classe di età', 
    y='Value', 
    hue='Tipo dato',
    palette='tab10',  # scegliamo una palette di colori diversa per le linee
    style='Tipo dato',  # aggiungiamo uno stile diverso per ogni tipo di dato
    markers=True,  # mostriamo i marker sui punti della linea
    dashes=False  # disabilitiamo le linee tratteggiate per rendere il grafico più nitido
)

# Miglioriamo l'estetica del grafico
plt.title('Tasso di Felicità della Vita per Fascia di Età nel 2022')
plt.xlabel('Fascia di Età')
plt.ylabel('Tasso di Felicità della Vita')

# Mostrare il grafico
plt.legend(loc='upper right', bbox_to_anchor=(1.25, 1))
plt.show()
```

### **Grafico di Dispersione: Correlazione tra "Felicità" e "Amici"**

#### Descrizione:
Il grafico di dispersione mostra la relazione tra la frequenza degli incontri con amici e il livello di felicità. Questo grafico rappresenta i dati per diverse classi di età, consentendo di visualizzare come la felicità varia in base al numero di incontri con amici.

#### Dettagli del Grafico:
- **Asse X (Orizzontale):** Numero di incontri con amici.
- **Asse Y (Verticale):** Livello di felicità.
- **Punti Dati:** Ogni punto rappresenta una combinazione di valori di "Amici" e "Felicità" per una specifica classe di età.

#### Interpretazione:
- **Pendenza della Linea di Regressione:** 
  - **Leggermente Positiva:** Conferma una correlazione leggermente positiva tra "Amici" e "Felicità", suggerendo che un aumento nel numero di incontri con amici è associato a un aumento del livello di felicità, sebbene la correlazione possa essere debole.


```python
df_amici = pd.read_csv('amici/amici2022.csv')
df_felicita = pd.read_csv('vita/vita2022.csv')

# Convertire la colonna "Value" in numerico per entrambi i dataset
df_amici['Value'] = pd.to_numeric(df_amici['Value'], errors='coerce')
df_felicita['Value'] = pd.to_numeric(df_felicita['Value'], errors='coerce')

# Rimuovere eventuali righe con valori mancanti
df_amici = df_amici.dropna(subset=['Value'])
df_felicita = df_felicita.dropna(subset=['Value'])

# Rinomininare la colonna "Value" in ciascun dataset per evitare conflitti dopo la fusione
df_amici = df_amici.rename(columns={'Value': 'Amici'})
df_felicita = df_felicita.rename(columns={'Value': 'Felicità'})

# Convertire la colonna "Classe di età" in numeri prendendo la media della fascia d'età
def convert_age_class(age_class):
    try:
        start, end = age_class.split('-')
        return (int(start) + int(end.split()[0])) / 2
    except ValueError:
        return None  # ritorna None se il formato non è corretto

df_amici['Classe di età'] = df_amici['Classe di età'].apply(convert_age_class)
df_felicita['Classe di età'] = df_felicita['Classe di età'].apply(convert_age_class)

# Rimuovere eventuali righe con valori mancanti dopo la conversione
df_amici = df_amici.dropna(subset=['Classe di età'])
df_felicita = df_felicita.dropna(subset=['Classe di età'])

# Unire i dataset sulla base della colonna 'Classe di età'
merged_df = pd.merge(df_amici, df_felicita, on='Classe di età')

# Selezionare solo le colonne numeriche per la correlazione
numerical_cols = ['Amici', 'Felicità']
merged_df = merged_df[numerical_cols]

# Calcolare la matrice di correlazione
correlation_matrix = merged_df.corr()

# Visualare la matrice di correlazione
print(correlation_matrix)

# Creare il grafico di correlazione
plt.figure(figsize=(14, 10))
correlation_plot = sns.lmplot(
    data=merged_df,
    x='Amici', 
    y='Felicità',
    palette='tab10',
    markers='o'
)

# Miglioriamo l'estetica del grafico
plt.title('Controllo correlazione tra Incontro con Amici e Felicità della Vita')
plt.xlabel('Incontro con Amici')
plt.ylabel('Felicità della Vita')
plt.grid(True)

# Mostriamo il grafico
plt.show()
```
